```python
import folium
import geopandas as gpd
import pandas as pd
```

```python
polbndry = gpd.read_file("Lab3_Dataset/polbndry.shp")
place = gpd.read_file("Lab3_Dataset/places.shp")
tambon = gpd.read_file("Lab3_Dataset/tambon.dbf",encoding='tis-620')
pl_type = gpd.read_file("Lab3_Dataset/PL_TYPE.dbf",encoding='tis-620')
landuse = gpd.read_file("Lab3_Dataset/landuse.shp")
```

```python
print(polbndry.crs)
print(place.crs)
print(landuse.crs)
```

```python
polbndry = polbndry.set_crs('EPSG:24047')
place = place.set_crs('EPSG:24047')
landuse = landuse.set_crs('EPSG:24047')
print(polbndry.crs)
print(place.crs)
print(landuse.crs)
```

```python
print(polbndry.columns)
print(tambon.columns)
print(place.columns)
print(pl_type.columns)
```

```python
polbndry = polbndry.merge(tambon, on = 'TAMBON_IDN', how = 'left')
place = place.merge(pl_type, on = 'PL_TYPE', how = 'left')
```
