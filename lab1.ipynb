```python
import geopandas as gpd
import matplotlib.pyplot as plt
import pandas as pd
import folium
import numpy as np
```

```python
gb_vol = gpd.read_file("World/Global_Volcano.shp")
w_city = gpd.read_file("World/World_City.shp")
w_country = gpd.read_file("World/World_Country.shp")
w_continent = gpd.read_file("World/World_Continent.shp")
w_rivers = gpd.read_file("World/World_River.shp")
w_mjr_lakes = gpd.read_file("World/World_MajorLake.shp")
```

```python
m = folium.Map(location=[0, 0], zoom_start=2)
folium.GeoJson(gb_vol,name="Volcanoe").add_to(m)
folium.GeoJson(w_rivers,name="River").add_to(m)
folium.GeoJson(w_mjr_lakes,name="Major Lakes").add_to(m)
folium.LayerControl().add_to(m)
m
```

```python
m = folium.Map(location=[0, 0], zoom_start=2)
folium.GeoJson(w_city,
               name="City",
               marker = folium.CircleMarker(
                   radius = 4,
                   color = 'black',
                   fill = True,
                   fill_color = 'red',
                   fill_opacity = 0.9
               )).add_to(m)
```

```python
color_dict = {
    "Africa": "purple",
    "Antarctica": "blue",
    "Asia": "cyan",
    "Australia": "green",
    "Europe": "lightgreen",
    "North America": "orange",
    "Oceania": "brown",
    "South America": "red"
}
def style_function(feature):
    cont = feature["properties"]["CONTINENT"]
    return {
        "fillColor": color_dict.get(cont, "gray"),
        "color": "black",      
        "weight": 1,
        "fillOpacity": 0.25 
    }
folium.GeoJson(
    w_continent,
    name="Continent",
    style_function=style_function
).add_to(m)
```

```python
folium.Choropleth(
    geo_data=w_country.to_json(),
    data=w_country,
    columns=["COUNTRY", "pop_2023"],
    key_on="feature.properties.COUNTRY",
    fill_color="PuRd",
    fill_opacity=0.7,
    line_opacity=0.5,
    nan_fill_color="gray",
    bins = 6,
    name="Population 2023"
).add_to(m)
```

```python
folium.LayerControl().add_to(m)
m.save("map1.html")
```
